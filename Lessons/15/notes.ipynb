# Problemy programisty

## Kod binarny

### Czym jest kod binarny?

Kod binarny to system zapisu informacji, w którym używa się tylko dwóch symboli: 0 i 1. Jest podstawą działania komputerów i elektroniki cyfrowej, ponieważ odpowiada dwóm stanom fizycznym – np. wyłączony/włączony, brak/napięcie.

| Liczba dziesiętna | Kod binarny |
|-------------------|-------------|
| 1                 | 1           |
| 2                 | 10          |
| 3                 | 11          |
| 4                 | 100         |
| 5                 | 101         |
| 6                 | 110         |
| 7                 | 111         |
| 8                 | 1000        |
| 9                 | 1001        |
| 10                | 1010        |

### Jak przeliczać z systemu dziesiętnego na binarny?

Jeśli liczba jest parzysta zapisujemy 0 jeśli nie to 1, następnie liczbę dzielimy przez 2 bez reszty i powtarzamy proces aż do momentu, gdy otrzymamy w wyniku dzielenia 0.
Wtedy odczytujemy zapisane liczby od końca.

$$

\begin{aligned}
41 | 1\text{, bo } 41 \% 2 = 1 \\
20 | 0 \\
10 | 0 \\
5 | 1 \\
2 | 0 \\
1 | 1 \\
0 \\
41\text{ (dziesiętnie)} = 101001\text{ (binarnie)}
\end{aligned}

$$

### Jak przeliczać z systemu binarnego na dziesiętny?

Aby obrócić proces musimy policzyć to na następującej zasadzie, każdą liczbę kodu
binarnego mnożymy przez 2 podniesione do potęgi pozycji (pamiętając, że liczymy od
zera).

Przykład dla liczby `101001`, czyli `41`.

| Bit  | Wartość potęgi 2 |
|------|------------------|
| 1    | 2^5              |
| 0    | 2^4              |
| 1    | 2^3              |
| 0    | 2^2              |
| 0    | 2^1              |
| 1    | 2^0              |

co daje nam

$1*2^5 + 0*2^4 + 1*2^3 + 0*2^2 + 0*2^1 + 1*2^0 = 41$

## Liczby pierwsze

Liczby pierwsze to liczby naturalne większe od 1, które mają dokładnie dwa dzielniki: 1 i samą siebie.
Znajdują zastosowanie m.in. w kryptografii (np. szyfrowanie RSA), teorii liczb oraz w algorytmach komputerowych, gdzie ich właściwości zapewniają bezpieczeństwo i efektywność obliczeń.

## Palindromy

Palindrom to wyraz, liczba, zdanie lub inny ciąg znaków, który czytany od lewej do prawej i od prawej do lewej brzmi tak samo.

Przykłady Palindromów:
- oko,
- sedes,
- zakopane na pokaz,
- elf układał kufle,
- 4554,
- kamil ślimak.

## Anagramy

Anagram to słowo lub wyrażenie utworzone przez przestawienie liter innego słowa lub wyrażenia, przy zachowaniu wszystkich oryginalnych znaków.

Przykłady anagramów:
- lama → mala,
- rak → kar,
- sen → nes,
- Roma → amor.
