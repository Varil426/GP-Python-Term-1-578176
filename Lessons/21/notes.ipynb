# Konstruktor

## `__init__`

```python
class Uczen:
    imie = ""
    nazwisko = ""
    licznik_uczniow = 0
    zmienna_powiazana_z_klasa = "TEST"
    
    def __init__(self, imie, nazwisko):
        self.imie = imie
        self.nazwisko = nazwisko
        Uczen.licznik_uczniow += 1
        # self.licznik_uczniow = -1
```

```python
uczen_1 = Uczen("Jan", "Kowalski")
uczen_2 = Uczen("Anna", "Nowak")
```

## Zmienne powiązane z klasą, a zmienne powiązane z obiektem

```python
print(Uczen.imie)
print(uczen_1.imie)
print(Uczen.licznik_uczniow)
print(uczen_1.licznik_uczniow)
```

```python
uczen_1.zmienna_powiazana_z_klasa = "TEST 123"
uczen_2.zmienna_powiazana_z_klasa = "TEST 321"

print(uczen_1.zmienna_powiazana_z_klasa)
print(uczen_2.zmienna_powiazana_z_klasa)
print(Uczen.zmienna_powiazana_z_klasa)
```
